## 2. Now write a function that returns the largest element on a list

```python
def greatest(input_list):
   

    largest = input_list[0]
    for number in input_list:
        if number > largest:
            largest = number
    return largest
print(greatest([1, 200, 3, 4, 5]))  # Output: 200
```

## 6. Write a function that returns the factorial of a number.

```python
def factorial(n):
#your code here
  if n < 0:
        return "Invalid input"
  elif n == 0 or n == 1:
        return 1
  else:
        result = 1
    
        for i in range(2, n + 1):
            result = result * i
        return result
print(factorial(5))  # Output: 120
print(factorial(1))  # Output: 1
print(factorial(0))  # Output: 1
print(factorial(-3))  # Output: Invalid input


    
```

```python
#factorial formula
#n! = n * ( n - 1 ) *...*1

# This code defines a function called "factorial" which takes an input "n". The function uses a for loop to iterate through the range of numbers 
# from 1 to n+1. For each number in that range, it multiplies the current value of x by the number in the range. At the end of the loop, 
# the function returns the final value of x, which will be the factorial of the input number "n".

# The Factorial of a positive integer n is the product of all positive integers less than or equal to n. 
# For example, the factorial of 6 (written "6!") is 6 * 5 * 4 * 3 * 2 * 1 = 720.

# So this function takes an input of any positive integer, and returns the factorial of that number.
```

```python
# This will test your function 
test_factorial(factorial)
```

## 7. Write a function that takes a list and returns a list of the unique values.

`NOTE: You cannot use set. 🤔`

```python
def unique(lst_un):
#your code here
    unique_list = []
    for item in lst_un:
        if item not in unique_list:
            unique_list.append(item)
    return unique_list
print(unique([1, 1, 1, 2, 3, 4, 5, 1, 2, 3, 3, 4, 5, 6, 7, 3, 3, 8, 1, 8, 9, 10]))
print(unique(['a', 'a', 'b', 'c', 'a', 'd', 'b', 'c', 'a', 'a', 'a',]))
```

## 8. Write a function that returns the mode of a list, i.e.: the element that appears the most times.
`NOTE: You should not use count... 🧐`

```python
def mode_counter(arr):
#your code here
    frequency = {}
    for item in arr:
        if item in frequency:
            frequency[item] += 1
        else:
            frequency[item] = 1
    mode_value = arr[0]   
    highest_count = 0
    
    for item, count in frequency.items():
        if count > highest_count:
            highest_count = count  # Update the record score
            mode_value = item      # Update the winning item
            
    return mode_value  # Return just the winner, not the whole dictionary
print(mode_counter([1, 1, 1, 1, 2, 4, 5, 1, 2, 3, 2, 2])) 
 

      
                    
    
```

## 9. Write a function that calculates the standard deviation of a list.
`NOTE: Do not use any libraries or already built functions. 😉`

```python
def st_dev(list_sd):
#your code here
    n = len(list_sd)
    if n <= 1:
        return 0.0

    mean = sum(list_sd) / n

    variance = sum((x - mean) ** 2 for x in list_sd) / (n - 1)
    std_dev = variance ** 0.5
    return std_dev
print(st_dev([1, 23, 333, 444, 3445]))
```

```python
# This will test your function 
test_stdev(st_dev)
```

## 11. Write a function that receives a string of comma separated words and returns a string of comma separated words sorted alphabetically.

`NOTE: You may use sorted but not split and definitely no join! 🤪`

```python
def sort_alpha(string):
    #  1: split the string by commas
    words_list = []
    current_word = ""
    
    for char in string:
        if char == ",":
            words_list.append(current_word)
            current_word = ""  
        else:
            current_word += char  
            
    
    words_list.append(current_word)
    
    # 2: Sort the list alphabetically
    sorted_list = sorted(words_list)
    
    #  3: rebuild the string 
    result_string = ""
    for word in sorted_list:
        result_string += word + ","
        
    # Chop off the very last extra comma using string slicing
    final_output = result_string[:-1]
    
    return final_output

# Testing the final code
print(sort_alpha("goal,banana,apple,cherry"))  
    
```

```python
# This will test your function 
test_alpha(sort_alpha)
```

## 12. Write a function to check if a given password is strong (at least 8 characters, at least one lower case, at least one upper case, at least one number and at least one special character). It should output True if strong and False if not.

```python
def check_pass(password):
    #  Length check
    if len(password) < 8:
        return False

    # Character checks 
    has_upper = any(char.isupper() for char in password)
    has_lower = any(char.islower() for char in password)
    has_digit = any(char.isdigit() for char in password)
    
    #  special character
    has_special = any(not char.isalnum() for char in password)

    # 3. Return True only if all conditions are met
    return has_upper and has_lower and has_digit and has_special

# Testing
print(check_pass("Password123"))   # Output: False (No special character)
print(check_pass("Password123!"))  # Output: True
```

```python
# This will test your function 
test_pass(check_pass)
```
